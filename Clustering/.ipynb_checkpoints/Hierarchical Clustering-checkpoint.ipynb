# Using Single Linkage 

```python
import numpy as np

from scipy.cluster.hierarchy import dendrogram, linkage
from scipy.spatial.distance import squareform

import matplotlib.pyplot as plt


mat = np.array([[0.0, 9.0, 3.0, 6.0, 11.0], [9.0, 0.0, 7.0, 5.0, 10.0], [3.0, 7.0, 0.0, 9.0, 2.0], [6.0, 5.0, 9.0, 0.0, 8.0], [11.0, 10.0, 2.0, 8.0, 0.0]])
dists = squareform(mat)
linkage_matrix = linkage(dists, "single")
dendrogram(linkage_matrix, labels=["A", "B", "C", "D", "E"])
plt.title("test")
plt.show()
```

# Using Complete Linkage 

```python
import numpy as np

from scipy.cluster.hierarchy import dendrogram, linkage
from scipy.spatial.distance import squareform

import matplotlib.pyplot as plt


mat = np.array([[0.0, 9.0, 3.0, 6.0, 11.0], [9.0, 0.0, 7.0, 5.0, 10.0], [3.0, 7.0, 0.0, 9.0, 2.0], [6.0, 5.0, 9.0, 0.0, 8.0], [11.0, 10.0, 2.0, 8.0, 0.0]])
dists = squareform(mat)
linkage_matrix = linkage(dists, "complete")
dendrogram(linkage_matrix, labels=["A", "B", "C", "D", "E"])
plt.title("test")
plt.show()
```

```python
import pandas as pd
from scipy.spatial import distance_matrix

data = [[2, 1], [1, 14], [10, 7], [1, 11], [3, 4], [11, 8], [4, 3], [12, 9]]
ctys = ['P1', 'P2', 'P3', 'P4', 'P5', 'P6','P7', 'P8']
df = pd.DataFrame(data, columns=['xcord', 'ycord'], index=ctys)
df
```

```python
distance_matrix = pd.DataFrame(distance_matrix(df.values, df.values), index=df.index, columns=df.index)
distance_matrix
```
